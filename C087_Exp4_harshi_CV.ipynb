```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_ori=cv2.imread('/content/text.png')
image_ori = cv2.cvtColor(image_ori, cv2.COLOR_BGR2RGB)
image=image_ori.copy()
```

```python

plt.imshow(image_ori)
```

```python
image_g = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_g, cmap='gray')
```

```python
R = cv2.cornerHarris(image_g, 3, 3, 0.04)
```

```python
image_g.shape
```

```python
th = 0.1*R.max()
```

```python
[rows,cols] = R.shape
```

```python
for r in range(0,rows):
  for c in range(0,cols):
    if R[r,c]>th:
      cv2.circle(image,(c,r),8,(255,0,0),5)# 4-radius,thickness-1
```

```python
plt.imshow(image)
```

```python
plt.subplot(1,2,1)
plt.imshow(image_ori)
plt.title('Original')
plt.subplot(1,2,2)
plt.imshow(image)
plt.title('Corners')
```

```python
th_negative = 0.1*R.min()
[rows,cols] = R.shape
```

```python
for r in range(0,rows):
  for c in range(0,cols):
    if R[r,c]<th_negative:
      cv2.circle(image,(c,r),3,(255,0,0),1)# 4-radius,thickness-1
```

```python
plt.imshow(image)
plt.title('Image with Edges')
```

```python
image=cv2.imread('/content/images.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image_g=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image)
```

```python
R = cv2.cornerHarris(image_g, 3, 3, 0.04)
[rows,cols] = R.shape
th = 0.1*R.max()
```

```python
for r in range(0,rows):
  for c in range(0,cols):
    if R[r,c]>th:
      cv2.circle(image,(c,r),3,(255,0,0),2)# 4-radius,thickness-1
```

```python
plt.imshow(image)
plt.title('Image with Corners')
```

```python
th_negative = 0.1*R.min()
[rows,cols] = R.shape
for r in range(0,rows):
  for c in range(0,cols):
    if R[r,c]<th_negative:
      cv2.circle(image,(c,r),1,(255,0,0),1)# 4-radius,thickness-1
```

```python
plt.imshow(image)
plt.title('Image with Edges')
```

Conclusion : Harris corner is used to detect the corner and edges on the given image text.png,for positive threshold = 10% of the maximum corner response, it shows corners of the letters.

For Negative threshold =10% of corner response it shows edges of the letters.

Same corners are detected even if the image is rotated by 90 degress, and the brightness is reduced by 20.

the same algorithm is applied to image square to detact corners of square image.
